## 1.2 Transaction data: feature engineering

```python
# Divide into 2 dataset for further analyse
transaction_approved = transaction_dropna[transaction_dropna['order_status'] == 'Approved']
transaction_declined = transaction_dropna[transaction_dropna['order_status'] != 'Approved']
```

```python
# Frequency (F) in RFM model
# No. of purchase
purchase_approved_count = transaction_approved.groupby('customer_id')['transaction_id'].count()
purchase_approved_count.name = 'purchase_approved_count'
purchase_declined_count = transaction_declined.groupby('customer_id')['transaction_id'].count()
purchase_declined_count.name = 'purchase_declined_count'
```

```python
# Recency (R) in RFM model
# Date of Recent purchase
recent_purchase_date = transaction_approved.groupby('customer_id')['transaction_date'].max()
least_recency = pd.Timestamp('2018-01-01 00:00:00') - recent_purchase_date
least_recency = least_recency.apply(lambda x:x.days)
least_recency.name = 'least_recency'
```

```python
# Monetary Value(M) in RFM model
# Total Value(price&cost) of approved orders
price_sum = transaction_approved.groupby('customer_id')['list_price'].sum()
price_sum.name = 'price_sum'
cost_sum = transaction_approved.groupby('customer_id')['standard_cost'].sum()
cost_sum.name = 'cost_sum'
```

```python
# percentage of online order
online_order_percentage = transaction_approved.groupby('customer_id')['online_order'].sum()/purchase_approved_count
online_order_percentage.name = 'online_order_percentage'
```

## 1.8 Further data engineering (data transforming) for ML model

```python
customer_attributes = test_clean.columns
transaction_features = train.columns.difference(customer_attributes)
```

```python
# Find out text features
text_features = []
for x in customer_attributes:
    if type(train[x][0]) == str:
        text_features.append(x)

        
text_features
```

```python
# Transform text features into numerical features
dict_list = []
from sklearn.preprocessing import LabelEncoder
for feature in text_features:
    encoder = LabelEncoder().fit(train[feature])
    dict_list.append(dict(zip(encoder.classes_,encoder.transform(encoder.classes_))))
    train[feature] = encoder.transform(train[feature])
    test_clean[feature] = encoder.transform(test_clean[feature])
```

```python
continuous_featuers = customer_attributes.difference(text_features)
continuous_featuers = continuous_featuers.drop(['age_cut','Mass Customer'])
```

```python
# Standardise countinuous features
standardise_dict = []
from sklearn.preprocessing import StandardScaler
for feature in continuous_featuers:
    scaler = StandardScaler().fit(np.array(train[feature]).reshape(-1, 1))
    train[feature] = scaler.transform(np.array(train[feature]).reshape(-1, 1))
    test_clean[feature] = scaler.transform(np.array(test_clean[feature]).reshape(-1, 1))
    standardise_dict.append((feature, scaler.mean_[0], scaler.var_[0]))
    
scaler = StandardScaler().fit(train[transaction_features])
train[transaction_features] = scaler.transform(train[transaction_features])
```

```python
standardise_df = pd.concat([pd.DataFrame((transaction_features,scaler.mean_,scaler.var_)).T,
                            pd.DataFrame(standardise_dict)],axis=0)
standardise_df.columns=[['feature_name','mean','var']]
standardise_df = standardise_df.reset_index(drop=True)
standardise_df
```

# 2. ML Model 1: identify high value existing customers through transaction data
Use K-Means model

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2).fit(train[transaction_features])
train['high_value'] = kmeans.labels_
```

```python
# Make sure that high value customers are labeled as 1
if kmeans.cluster_centers_[0][0] > kmeans.cluster_centers_[1][0]: train['high_value'] = 1 - train['high_value']
```

# 3. ML Model 2: Try to identify high value new customers through demographic attributes only
**Background: New customers havn't done any transaction**  
**Dependent variable: labels from K-means,  Independent variables: demographic attributes**  
First, find out waht demographic attributes may affect the target(EDA)

```python
train_labeled = pd.concat([train_origin,train['high_value']],axis=1)
```

```python
customer_attributes
```

```python
sns.heatmap(train_labeled[['high_value','past_3_years_bike_related_purchases','tenure','property_valuation']].corr(),cmap='RdBu',annot=True)
plt.title('Correlation coefficient between continuous variables and target(high value classification)',)
```

```python
def feature_groupby_info(feature):
    groupby_feature = train_labeled.groupby([feature,'high_value'])['least_recency'].count().unstack()
    groupby_feature['high_value_rate'] = groupby_feature.apply(lambda x:x[1]/sum(x),axis=1)
    print('Variance of "high value rate":',groupby_feature['high_value_rate'].var())
    return(groupby_feature.sort_values('high_value_rate',ascending=False))
```

```python
y = train.high_value
X = train[customer_attributes]
predict_X = test_clean[customer_attributes]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
randomforest = RandomForestClassifier(n_estimators=5,max_depth=2)
param_grid = {'max_leaf_nodes':[2,3,4,5,6,7,8,9]}
gridsearch = GridSearchCV(randomforest, param_grid=param_grid, cv= 5, n_jobs=-1, return_train_score=True)
gridsearch.fit(X_train,y_train)
gridsearch.cv_results_
```

```python
print('roc_auc scores: ')
print('train set: ',roc_auc_score(y_train,gridsearch.predict(X_train)))
print('test set: ',roc_auc_score(y_test,gridsearch.predict(X_test)))
```

```python
randomforest = RandomForestClassifier(n_estimators=5,max_depth=2,max_leaf_nodes=6)
randomforest.fit(X_train,y_train)
print('roc_auc scores: ')
print('train set: ',roc_auc_score(y_train,randomforest.predict(X_train)))
print('test set: ',roc_auc_score(y_test,randomforest.predict(X_test)))
```

Shows no clear difference from random classifier
