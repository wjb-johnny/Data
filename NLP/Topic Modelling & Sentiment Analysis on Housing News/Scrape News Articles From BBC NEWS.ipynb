# Scrape News Articles From BBC NEWS

```python
# Import packages for scraping
#https://sites.google.com/a/chromium.org/chromedriver/
from selenium.webdriver import Chrome
from datetime import datetime as dt
import time
import pandas as pd
```

```python
# Check if we are allowed to scrape:
# https://www.promptcloud.com/blog/how-to-read-and-respect-robots-file/
# https://www.bbc.co.uk/robots.txt
# Set url
url = "https://www.bbc.co.uk/news/topics/cdl8n2edgj5t/housing-market"
# Set the web driver path (https://sites.google.com/a/chromium.org/chromedriver/downloads)
webdriver = "C:/Users/user/AppData/Local/chromedriver_win32/chromedriver.exe"
# Set the driver
driver = Chrome(webdriver)
driver.get(url)
# Find the button
button = driver.find_element_by_xpath("//button[@class='lx-stream-show-more__button gel-long-primer-bold']")
# Press the button a couple of times
for j in range(15):
    try:
        time.sleep(20)
        button.click()
    except:
        print("Failed to press button at: ", j)
        pass
```

```python
def extract_date(date_list):
    # today's news(no date/year): date_list = [[],[]]
    # this year's news(no year, only have date): date_list = [[WebElement],[]]
    # previous news(have both): date_list = [[WebElement],[WebElement]]
    if not date_list[0]: return dt.today() #today 
    elif not date_list[1]: return dt.strptime(date_list[0].text+' '+str(dt.today().year),'%d %b %Y') # this year
    else: return dt.strptime(date_list[0].text+' '+date_list[1][0].text,'%d %b %Y') #previous
```

```python
# Sleep it a bit
# time.sleep(5)
# Let's start collecting articles/headlines + datetime
headlines = driver.find_elements_by_class_name("gs-o-media__body") 
date_time = driver.find_elements_by_class_name('lx-stream-post__meta')
total = []
ind_time = 0
for j in range(len(headlines)):
    try:
        headline_text = headlines[j].find_element_by_class_name('lx-stream-post__header-text').text
        date_text_raw = [date_time[ind_time].find_element_by_class_name("qa-meta-date")]
        date_text_raw.append(date_time[ind_time].find_elements_by_class_name("qa-meta-year"))
        ind_time += 1
        date_text = extract_date(date_text_raw)
        new = ((date_text, headline_text))
        print(new)
        total.append(new)
    except:
        pass
# Collect in a dataframe
df = pd.DataFrame(total,columns=['date', 'headline'])
# Print it to csv
# df.to_csv('Headlines_HMUK.csv')
# Close connection
#driver.close()
```
